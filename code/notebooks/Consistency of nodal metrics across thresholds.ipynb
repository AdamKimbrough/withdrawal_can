```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
from scipy import stats
%matplotlib inline
```

## Comparison across thresholds.

## Louvain (thresholded)

```python
netdata_thresh_085 = pd.read_csv('/Users/dlurie/Dropbox/Projects/withdrawal_can/results/withdrawal_thresh085/gamma1.0_combined_results.csv', index_col='Region')
netdata_thresh_075 = pd.read_csv('/Users/dlurie/Dropbox/Projects/withdrawal_can/results/withdrawal_thresh075/gamma1.0_combined_results.csv', index_col='Region')
```

#### PC

```python
sns.regplot(netdata_thresh_075.pc_pos, netdata_thresh_085.pc_pos)
plt.xlabel('Low Threshold')
plt.ylabel('Medium Threshold')
stats.spearmanr(netdata_thresh_075.pc_pos, netdata_thresh_085.pc_pos)
```

What percentage of the 15 regions with the highest PC replicate across thresholds?

```python
(len(set.intersection(set(netdata_thresh_075.pc_pos.nlargest(15).index), set(netdata_thresh_085.pc_pos.nlargest(15).index)))/15)*100
```

#### WMDz

```python
sns.regplot(netdata_thresh_075.wmdz_pos, netdata_thresh_085.wmdz_pos)
plt.xlabel('Low Threshold')
plt.ylabel('Medium Threshold')
stats.spearmanr(netdata_thresh_075.wmdz_pos, netdata_thresh_085.wmdz_pos)
```
