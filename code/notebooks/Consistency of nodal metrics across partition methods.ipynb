```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
netdata_thresh = pd.read_csv('/Users/dlurie/Dropbox/Projects/withdrawal_can/results/withdrawal_thresh075/gamma1.0_combined_results.csv', index_col='Region')
netdata_nothresh = pd.read_csv('/Users/dlurie/Dropbox/Projects/withdrawal_can/results/withdrawal_thresh075/NothreshPartition_combined_results.csv',index_col='Region')
netdata_hac = pd.read_csv('/Users/dlurie/Dropbox/Projects/withdrawal_can/results/withdrawal_thresh075/HACPartition_combined_results.csv',index_col='Region')
```

### Compare PC across partitions
Node color determined by PC of nodes in the HAC partition.

```python
plt.scatter(netdata_thresh.pc_pos, netdata_nothresh.pc_pos, c=netdata_hac.pc_pos)
plt.xlabel("Thresholded Louvain")
plt.ylabel("Unthresholded Louvain")
plt.colorbar()
```

```python
pc_df = pd.DataFrame(np.array([netdata_thresh.pc_pos, netdata_nothresh.pc_pos, netdata_hac.pc_pos]).T,
                   columns=['louvain_thresh', 'louvain_nothresh', 'hac'])
```

```python
sns.pairplot(pc_df)
```
